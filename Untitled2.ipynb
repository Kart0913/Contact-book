```python
"""
CONTACT BOOK PROJECT
"""
# ---------- STEP 1: LOGIN CHECK  ----------
owner_name = "keerti"   # <-- change this to your actual name/username

entered_name = input("Enter your username to access the Contact Book: ")

if entered_name == owner_name:
    print("\nAccess Granted! Welcome,", owner_name)

    # ---------- STEP 2: CONTACT BOOK DATA ----------
    contacts = {}   

    # ---------- STEP 3: MAIN MENU LOOP ----------
    running = True
    while running:
        print("\n----- CONTACT BOOK MENU -----")
        print("1. Add Contact")
        print("2. Search Contact")
        print("3. Delete Contact")
        print("4. Show All Contacts")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        # ---- ADD CONTACT ----
        if choice == "1":
            name = input("Enter name: ")
            phone = input("Enter phone number: ")
            contacts[name] = phone
            print(name, "added successfully!")

        # ---- SEARCH CONTACT ----
        elif choice == "2":
            search_name = input("Enter name to search: ")
            if search_name in contacts:
                print("Name :", search_name)
                print("Phone:", contacts[search_name])
            else:
                print(search_name, "not found in contact book.")

        # ---- DELETE CONTACT ----
        elif choice == "3":
            del_name = input("Enter name to delete: ")
            if del_name in contacts:
                del contacts[del_name]
                print(del_name, "deleted successfully!")
            else:
                print(del_name, "not found in contact book.")

        # ---- Display all contacts ---
        elif choice == "4":
            if len(contacts) == 0:
                print("Contact book is empty.")
            else:
                print("\nAll Contacts:")
                names_list = list(contacts.keys())
                i = 0
                while i < len(names_list):
                    current_name = names_list[i]
                    print("Name :", current_name)
                    print("Phone:", contacts[current_name])
                    print("-" * 20)
                    i += 1

        # ---- EXIT ----
        elif choice == "5":
            print("Exiting Contact Book. Goodbye!")
            running = False

        # ---- INVALID CHOICE ----
        else:
            print("Invalid choice. Please enter 1-5.")

else:
    print("\nAccess Denied! This is not your contact book.")
```
